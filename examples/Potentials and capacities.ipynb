## Stack design

This example notebook shows how to supply parameters to a TEA class, how to plot mass- and volume loadings, how to estimate energy densities and how to loop over different parameter settings.

### Let's generate TEA classes and loading plots

A good way to create an electrode stack cell parameter-set is to start with a base parameter-set, containing the desired active materials, and update parameter such as densities with inputs for the TEA class.

It is also possible to create parameter-sets from the bottom-up, e.g. for non-lithium chemistries, however then the key benefit of the library: PyBaMM may not work with the parameter-set. PyBaMM parameter-sets can also directly be used, for the loadings plot usally active material- and electrolyte densities are necessary inputs though.

Below one "design" input-set, one "chemistry" input-set and one "only_electrolyte_density" input-set are defined to create four TEA classes, one with the "Chen2020" parameter-set, one without a base parameter-set and two classes having the same "design" input-set, but different active material related base parameter-sets.

```python
#%pip install pybamm
import pybamm
from pybamm_tea import tea
import numpy as np
import matplotlib.pyplot as plt

### create input-sets

# input only
only_electrolyte_density = {
    "Electrolyte density [kg.m-3]": 1276, # LiPF6 in EC:EMC 3:7 + 2% VC
#    "Separator material density [kg.m-3]": 946, # PP/PE

}

# stack design parameter
design = {
    # Negative electrode
    "Theoretical n/p ratio": 0.7, # instead of passing a thickness, one can also pass the ratio of theoretical maximal lithium content from the active material particles in both electrodes - not considering potential/stoichiometry windows or lithium/active material losses
    "Negative electrode porosity": 0.33,
    "Negative electrode binder density [kg.m-3]": (2 * 940 + 1600) / 3, # 2:1 CMC:SBR
    "Negative electrode conductive additive density [kg.m-3]": 1800, # CB
    "Negative electrode binder dry mass fraction": 0.03,
    "Negative electrode conductive additive dry mass fraction": 0.02,
    
    # Separator and electrolyte
    "Separator thickness [m]": 25e-6,
    "Separator porosity": 0.45,
    "Separator material density [kg.m-3]": 946, # PP/PE

    # Positive electrode    
    "Positive electrode thickness [m]": 80e-6,
    "Positive electrode porosity": 0.33,
    "Positive electrode inactive material density [kg.m-3]": 1800, # PVDF and CB
    "Positive electrode active material dry mass fraction": 0.95,

    # Current collectors
    "Negative current collector thickness [m]": 12e-6,
    "Negative current collector density [kg.m-3]": 8960, # Cu
    "Positive current collector thickness [m]": 16e-6,
    "Positive current collector density [kg.m-3]": 2700, # Al
}

# chemistry input without base parameter-set
chemistry = {    
    # Negative electrode - Graphite
    "Negative electrode active material density [kg.m-3]": 2266, # Graphite
    "Negative electrode active material capacity [mA.h.g-1]": 372, # refers to maximal concentration
    "Negative electrode minimum stoichiometry": 0.02,
    "Negative electrode maximum stoichiometry": 0.90,
    "Negative electrode average OCP [V]": 0.1, # vs Li/Li+ | dependent on ocv curve if not set

    # Positive electrode - NMC604
    "Positive electrode active material density [kg.m-3]": 4750,
    "Positive electrode active material capacity [mA.h.g-1]": 275, # = 96485 / 3.6 / (6.9 + 0.6 * 58.7 + 0 * 54.9 + 0.4 * 58.9 + 2 * 16) | q [mAh.g-1] = F [C.mol-1] / 3.6 [C.mAh-1] / (M_Li+x*M_Ni+y*M_Mn+z*M_Co+2*M_O) [g.mol-1]
    "Positive electrode minimum stoichiometry": 0.25,
    "Positive electrode maximum stoichiometry": 0.90,
    "Positive electrode average OCP [V]" : 3.8, # vs Li/Li+ | dependent on ocv curve if not set

    # Electrolyte
    "Electrolyte density [kg.m-3]": 1276, # LiPF6 in EC:EMC 3:7
}

# merge the design and chemistry dictionaries into a new dictionary that will be used as input for the TEA class
input_only = design.copy() # create a copy of the design dictionary
input_only.update(chemistry) # merge the chemistry dictionary into the new dictionary

# graphite||nmc chemistry input with a base parameter-set
c_nmc = {
    "Negative electrode active material density [kg.m-3]": 2266, # Graphite
    "Positive electrode active material density [kg.m-3]": 4750, # NMC
    "Electrolyte density [kg.m-3]": 1276, # LiPF6 in EC:EMC 3:7
}
c_nmc.update(design) # merge the design dictionary with nmc-chemistry input for the base parameter-set

# graphite||lfp chemistry input with a base parameter-set
c_lfp = {
    "Negative electrode active material density [kg.m-3]": 2266, # Graphite
    "Positive electrode active material density [kg.m-3]": 3600, # LFP
    "Electrolyte density [kg.m-3]": 1276, # LiPF6 in EC:EMC 3:7
}
c_lfp.update(design) # merge the design dictionary with the lfp-chemistry input for the base parameter-set

# base parameter-sets
param_chen = pybamm.ParameterValues("Chen2020")
param_c_lfp = pybamm.ParameterValues("Prada2013")

### create TEA classes and plot stack (-loading) breakdown
# create a TEA class with the input_only input-set and plot stack breakdown
print("Input_only")
tea_input_only = tea.TEA(input_only)
tea_input_only.plot_stack_breakdown()
print(tea_input_only.stack_breakdown)

# create a TEA class for the lfp input-set with the "Prada2013" parameter-set and plot stack breakdown
print("LFP")
tea_lfp = tea.TEA(param_c_lfp, c_lfp)
tea_lfp.plot_stack_breakdown()
print(tea_lfp.stack_breakdown)

# create a TEA class for the nmc input-set with the "Mohtat2020" parameter-set and plot stack breakdown
print("NMC")
tea_nmc = tea.TEA(param_chen, c_nmc)
tea_nmc.plot_stack_breakdown()
print(tea_nmc.stack_breakdown)

# create a TEA class for the "Chen2020" parameter-set and plot stack breakdown
print("Chen2020")
tea_chen = tea.TEA(param_chen, only_electrolyte_density)
tea_chen.plot_stack_breakdown()
tea_chen.stack_breakdown_dataframe
```


### Did you spot the negative separator material density?

The inactive/separator material density is calculated as the remainder from the (parameter-set-) electrode/separator density and the (input-) active material & electrolyte density with respect to the volume fractions.
\begin{equation}
\rho_{\text {inactive/separator }}=\frac{\rho_{\text {electrode/separator }}-v_{\text {active }} \rho_{\text {active }}-\varepsilon_{}\rho_{\text {electrolyte }}}{v_{\text {inactive/separator }}}
\end{equation}

In this specific example case, the mass loading from the electrolyte (blue) is larger than the mass loading from the entire wet separator (dashed line) leading to the negative separator material density. Another special case here is that there are no inactive materials considered explicitly, so that no active material densities need to be provided, as they can be calculated as the dry electrode densities.

### More densities

Let's now see how the energy densities of the Graphite||LFP and Graphite||NMC classes increase with the positive electrode thickness.
Note, that the stack-level volumetric energy density is calculated without considering transport and reaction overpotentials / area-specific impedances, as below.

\begin{equation}
E_{\text {stack }}^{\text {volumetric }}=\frac{\min \left(q_{\text {area }}^{+}, q_{\text {area }}^{-}\right)\left(\bar{U}_{+}-\bar{U}_{-}\right)}{\frac{d_{c c}^{-}}{2}+d_{-}+d_{\text {separator }}+d_{+}+\frac{d_{c c}^{+}}{2}}
\end{equation}\tag{2}

The gravimetric energy density is then obtained by dividing with the stack density. Without the to-be-implemented form factor calculation (depending on electrode areas and so on) cell-level energy densities could be estimated with given constant form factors.

```python
# generate an array of thicknesses from 10e-6 to 100e-6
thicknesses = np.linspace(20e-6, 120e-6, 11)
# initialize lists for the stack energy densities
stack_volumetric_energy_densities_nmc = []
stack_volumetric_energy_densities_lfp = []
stack_gravimetric_energy_densities_nmc = []
stack_gravimetric_energy_densities_lfp = []

c_nmc.update({"Positive electrode thickness [m]": None}) # otherwise the default thicknesses are is used and the NPR ratio is overwritten
c_lfp.update({"Positive electrode thickness [m]":  None})

# loop over the thicknesses and calculate the stack energy densities for the nmc and lfp input-sets
for thickness in thicknesses:
    # update the positive electrode thickness in the input-sets
    c_lfp.update({"Negative electrode thickness [m]": thickness})
    c_nmc.update({"Negative electrode thickness [m]": thickness})
    
    # create TEA classes with the updated input-sets
    tea_lfp = tea.TEA(param_c_lfp, c_lfp)
    tea_nmc = tea.TEA(param_chen, c_nmc)

    # save the calculated the stack energy densities in the lists
    stack_volumetric_energy_densities_nmc.append(tea_nmc.stack_energy_densities.get("Volumetric stack energy density [Wh.L-1]"))
    stack_volumetric_energy_densities_lfp.append(tea_lfp.stack_energy_densities.get("Volumetric stack energy density [Wh.L-1]"))
    stack_gravimetric_energy_densities_nmc.append(tea_nmc.stack_energy_densities.get("Gravimetric stack energy density [Wh.kg-1]"))
    stack_gravimetric_energy_densities_lfp.append(tea_lfp.stack_energy_densities.get("Gravimetric stack energy density [Wh.kg-1]"))
    
# plot stack energy density vs thickness
plt.figure()
plt.plot(thicknesses*1e6, stack_volumetric_energy_densities_lfp, label="LFP - volumetric", color='red')
plt.plot(thicknesses*1e6, stack_gravimetric_energy_densities_lfp, label="LFP - gravimetric", color='red', linestyle='dashed')
plt.plot(thicknesses*1e6, stack_volumetric_energy_densities_nmc, label="NMC - volumetric", color='black')
plt.plot(thicknesses*1e6, stack_gravimetric_energy_densities_nmc, label="NMC - gravimetric", color='black', linestyle='dashed')
plt.xlabel("Positive electrode thickness [um]")
plt.ylabel("Stack energy densities [Wh.L-1] and [Wh.kg-1]")
plt.legend()
plt.show()

# print stack energy densities data for last loop
print("NMC stack energy densities dict:")
print(tea_nmc.stack_energy_densities)
print("LFP stack energy densities dict:")
print(tea_lfp.stack_energy_densities)
```
